## 1. What Is an Object?

In Python, an **object** is a bundle of related data and functionality kept together in one place.

The good news: you have already been using objects this whole time. Strings, lists, and dictionaries are all objects that Python gives you out of the box.

What makes an object useful is that it combines two things:

---

### Attributes — *what an object knows*

An **attribute** is a piece of information (data) stored inside the object.

For example, a list object keeps track of how many items it contains. You can access that information directly from the object:

```
my_list = [10, 20, 30]
my_list.__len__()   # the list "knows" it has 3 items
```

---

### Behaviors (Methods) — *what an object can do*

A **behavior** (called a **method** in Python) is a function that belongs to the object. You call it using a dot `.` after the object's name.

For example, a list object can sort itself, append new items, or remove items — those are all behaviors built into every list.

```
my_list.append(40)  # the list can add a new item to itself
my_list.sort()      # the list can sort itself
```

---

The dot `.` is how you reach inside an object to grab an attribute or call a behavior. In the code cell below you will see this with a few of Python's built-in object types.

````python
# ── String object ─────────────────────────────────────────────────────────────

greeting = "hello world"

# BEHAVIOR: strings can uppercase themselves
print(greeting.upper())

# BEHAVIOR: strings can check what they start with
print(greeting.startswith("hello"))

# BEHAVIOR: strings can replace parts of themselves
print(greeting.replace("world", "Python"))
````

````python
# ── List object ───────────────────────────────────────────────────────────────

scores = [84, 91, 73, 88]

# BEHAVIOR: lists can append a new item
scores.append(95)
print(scores)

# BEHAVIOR: lists can sort themselves
scores.sort()
print(scores)

# BEHAVIOR: lists can count occurrences of a value
print(scores.count(95))
````

````python
# ── Dictionary object ──────────────────────────────────────────────────────────

student = {"name": "Alex", "grade": "A", "age": 17}

# BEHAVIOR: dictionaries can return all their keys
print(student.keys())

# BEHAVIOR: dictionaries can return all their values
print(student.values())

# BEHAVIOR: dictionaries can look up a value safely with a default fallback
print(student.get("school", "unknown"))
````

## 2. Abstraction

**Abstraction** is one of the core ideas in computer science. It means taking something complex and representing it in a simpler, more manageable way — hiding the messy details so you can think about the *thing as a whole* instead of all its individual pieces.

You use abstraction every day without thinking about it. When you say "I drove to school," you are not describing every gear shift, brake press, and turn signal. You abstracted all of that away into one idea: *driving*.

In programming, abstraction means grouping related data together under one name so the code reflects how you actually think about the world.

---

### The Problem: Treating a Student as Separate Variables

Imagine you are trying to work with student information in your program. A student has:
- a first name
- a track they are enrolled in
- a list of programming languages they are already familiar with

In procedural code (the style you have been writing so far), you might reach for three separate variables:

````python
fname     = "Alex"
track     = "Data Analytics"
tools = ["Excel", "SQL"]
````

This works fine for one student. But notice the problem — nothing in the code actually connects these three variables to each other. Python has no idea that `fname`, `track`, and `familiars` all belong to the same person. They are just three floating pieces of data sitting in memory.

Now imagine you have a second student:

````python
fname     = "Bob"
track     = "Data Analytics"
tools = ["Python", "SQL"]
````

You just overwrote the first student. To keep both, you would need to start inventing naming conventions like `student1_fname`, `student2_fname`, `student1_track` — and it only gets worse from there.

This is exactly the problem that **abstraction** solves. Instead of scattering a student's data across multiple unrelated variables, we want a way to bundle all of it together under one idea: a *Student*. That bundle is what an object gives us, and in the next section we will see how to build one.

## 3. Defining Your Own Class

Just like we can define our own functions in Python using the `def` keyword, we can also define our own **classes** — the blueprints that describe what an object looks like and what it can do. Once a class is defined, we can create as many objects from it as we need, and each one will carry its own independent set of attributes and behaviors.

---

### The `class` keyword

You start a class definition with the `class` keyword, followed by the name of the class and a colon. By convention, class names in Python are written in **PascalCase** — every word capitalized, no underscores.

```python
class Fellow:
    ...
```

Everything indented underneath the `class` line belongs to that class.

---

### The `__init__` method

The first thing you define inside a class is a special method called `__init__` (short for *initialize*). Python calls this method **automatically** the moment you create a new object from the class. Its job is to set up the object's attributes — the pieces of data it will carry with it.

````python
class Fellow:
    # dunder methods (you do not call __init__ yourself, but this allows you to call)
    def __init__(self, fname, track, tools):
        self.fname = fname
        self.track = track
        self.tools = tools
    
    # method that can only be called from our instance "object"
    def get_name(self):
        return self.fname
````

Here is a breakdown of every piece of that syntax:

| Piece | What it means |
|---|---|
| `def __init__` | Defines the initialization method — Python runs this automatically when a new object is created |
| `self` | Refers to the specific object being created — it is always the first parameter and Python passes it in for you |
| `fname, track, familiars` | The values you supply when creating the object |
| `self.fname = fname` | Permanently attaches the value to this object as an attribute |

---

### Creating an object from the class

Once the class is defined, you create an object by calling the class name like a function and passing in the required values. Python takes those values, runs `__init__`, and hands you back a fully set-up object.

````python
# We're calling Fellow as if its a function "Fellow()"
# But what's really going on here is Fellow.__init__(fname=..., track=..., tool=...)

person1 = Fellow("Farukh", "DA", ["Python"])
person2 = Fellow("Gaurav", "DA", ["Tableau"])
person3 = Fellow("Anil", "DA", ["React"])
````

````python
print(person2.fname)
print(person2.track)
print(person2.tools)
````

Now `person1`, `person2`, and `person3` are three completely separate objects sitting in memory at the same time. Changing one has no effect on the other.

## 4. Defining Functions Inside a Class (Methods)

We just learned how to give an object its attributes using `__init__`. Now we can also give it **behaviors** — things the object can do. In Python, a function defined inside a class is called a **method**.

Writing a method looks almost identical to writing a regular function. The only difference is:

1. It is indented inside the class body
2. Its first parameter is always **`self`**

---

### Why `self` again?

When you call a method on an object, Python needs to know *which* object is doing the action. `self` is how the method gets access to that specific object's attributes.

Think of it this way — if you have two `Fellow` objects and you call `.official_title()` on one of them, `self` is what tells the method "use *this* fellow's name and track, not the other one's."

You never pass `self` in yourself — Python handles that automatically when you use the dot notation.

```python
fellow1.official_title()   # Python passes fellow1 in as self behind the scenes
```

````python
class Fellow:

    def __init__(self, fname, track, tools):
        self.fname     = fname 
        self.track     = track  
        self.tools = tools  

    # BEHAVIOR: produce an official introduction string for this fellow
    def official_title(self):
        return self.fname + " from " + self.track + " knows the following tools: " + str(self.tools)
````

````python
person1 = Fellow("Farukh", "DA", ["Python"])
person2 = Fellow("Gaurav", "DA", ["Tableau"])
person3 = Fellow("Anil", "DA", ["React"])

print(person2.official_title())
print(person3.official_title())
````

## 5. Inheritance and Polymorphism

So far, every class we have built has started from scratch. But one of the most powerful ideas in object-oriented design is that **classes can be built on top of other classes**. This is called **inheritance**, and it is the main tool we use to avoid repeating ourselves when designing systems.

---

### Inheritance — *building on what already exists*

**Inheritance** lets you create a new class that automatically receives all the attributes and methods of an existing class. The existing class is called the **parent** (or base) class. The new class is called the **child** (or subclass).

Think of it like a template hierarchy. You define the shared, general behavior once in the parent, then each child class only adds or changes what makes it unique.

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return self.name + " makes a sound"


class Dog(Animal):          # Dog inherits everything from Animal
    def speak(self):        # Dog overrides the speak method with its own version
        return self.name + " barks"
```

In the example above, `Dog` does not need to redefine `__init__` — it inherits it from `Animal` for free. The only thing `Dog` does is replace the `speak` behavior with something more specific.

This keeps your code **DRY** (Don't Repeat Yourself). If ten different animal classes all share the same initialization logic, you write it once in the parent and every child gets it automatically.

---

### Polymorphism — *one interface, many forms*

**Polymorphism** (from the Greek for "many shapes") is what happens when different classes share the same method name but each one implements it differently.

The power of polymorphism is that code which calls `speak()` does not need to know whether it is dealing with a `Dog`, a `Cat`, or a `Bird`. It just calls the method, and each object responds in its own way.

```python
class Cat(Animal):
    def speak(self):
        return self.name + " meows"

class Bird(Animal):
    def speak(self):
        return self.name + " chirps"

animals = [Dog("Rex"), Cat("Luna"), Bird("Tweety")]

for animal in animals:
    print(animal.speak())   # each object responds differently to the same call
```

This is polymorphism in action. The loop does not need separate `if` statements for each animal type — it calls `speak()` uniformly and each class handles it in its own way.

---

### Why this matters for system design

When you design a system using inheritance and polymorphism together, you get two major benefits:

1. **Reusability** — shared logic lives in the parent class once, and every child inherits it without duplication. Adding a new type means writing only what is genuinely new.

2. **Extensibility** — you can add new child classes later without changing any of the existing code that uses them. As long as the new class follows the same interface (same method names), everything continues to work.

These two ideas — write shared behavior once, let each subclass specialize — are the foundation of most real-world object-oriented system design.
